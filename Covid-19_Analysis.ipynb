<h4>Remove all the records where 'Confirmed' cases is less than 10

```python
df = df[~(df.Confirmed<10)]
```

```python
df.head(30)
```

<h4>In which region maximum number of confirmed cases were recorded [top 5 region]

```python
df.groupby('Region')['Confirmed'].sum().sort_values(ascending=False).head(5)
```

<h4>In which region minimum number of Deaths cases were recorded

```python
df.groupby('Region')['Deaths'].sum().sort_values(ascending=True).head(20)
```
